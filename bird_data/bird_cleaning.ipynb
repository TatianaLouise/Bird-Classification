First clean the birds_df to show only the common names (labels) and scientific names of the birds.

```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Create variable for the birds.csv file
birds_file = Path('../bird_data/birds.csv')
```

```python
# Read csv into pandas df

birds_df = pd.read_csv(birds_file)
```

```python
# Preview the birds_df
birds_df.head()
```

```python
condition = birds_df['data set'] == 'train'
```

```python
mask = condition
clean_birds_df = birds_df[mask]

clean_birds_df
```

```python
# Convert to a csv file
csv_file_path = 'data/bird_data_cleaning/clean_birds.csv'
clean_birds_df.to_csv(csv_file_path, index = False)
```
